```python
import cv2
import glob
import os
import numpy as np
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') # import this from github
fullimgpath = glob.glob(r"C:\Users\dell\Desktop\saugat\own-data\BISHALPROJECT\BISHALPROJECT\*.jpg") #collect all pic from this path
```

```python
a=300
b=11
from cv2 import cvtColor
for er in fullimgpath:
    img = cv2.imread(er)
    img=np.array(img)
    img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY) #gray scale conversion
    aa,bb= img.shape
    #img = cv2.imread(er, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img,(int(bb/2),int(aa/2)))
    img = cv2.GaussianBlur(img, (5, 5), 1.3)  # gaussin blur
    
    faces = face_cascade.detectMultiScale(img)
    for (x, y, w, h) in faces:
        img = cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
        roi_color = img[y:y+h, x:x+w]
        #print("face found and save localy")
        path = r'C:\Users\dell\Desktop\saugat\own-data\face-bishal'   # cropped image are found here
        #save image as png
        roi_color = cv2.resize(roi_color,(200,200))# w,h
        cv2.imwrite( os.path.join(path,''+str(a) + str(b) + '_faces.jpg'), roi_color)
        print(a)
        #cv2.imshow('img', roi_color)
        #cv2.waitKey(100)
        a=a-1
        b=b+1
```
